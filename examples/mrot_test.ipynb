```python
from mrot import Rotator
from math import cos, sin, radians
import numpy as np
from skimage import data
from skimage import img_as_float
import mpes.visualization as vis
%matplotlib inline
```

## Display original image

```python
img = img_as_float(data.coins())
vis.colormesh2d(img.T, figsize=(6,8), colormap='Greys_r')
```

## Rotation about (0, 0)

```python
imgrotator = Rotator(img.T)
imgrotator.rotate(-20)
vis.colormesh2d(imgrotator.rotimg, figsize=(6,8), colormap='Greys_r')
```

## Rotate about an arbitrary origin

```python
imgrotator.rotate(-40, origin=(150,200))
vis.colormesh2d(imgrotator.rotimg, figsize=(6,8), colormap='Greys_r')
```
